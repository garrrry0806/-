```
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt

table = {
    "類別":[],
	"名次":[],
	"股號":[],
	"股名":[],
	"漲跌幅":[],
	"成交量":[]
}
```


# 定義函式： chip

```
def chip(url, headers, title):
    response = requests.get(url, verify=False, headers=headers)
    rawdata = json.loads(response.text)
    data = rawdata["Data"]["QueryStockResult"]

    print(title)
    # 名次、股號股名、漲跌幅、成交量
    for i in range(0, 15):
        print("名次", i + 1)
        print("股號：", data[i]["StockID"])
        print("股名：", data[i]["StockName"])
        print("漲跌幅：", data[i]["UpRate"])
        print("成交量：", data[i]["DealVol"])
        print("-" * 20)

        table["類別"].append(title)
        table["名次"].append(i + 1)
        table["股號"].append(data[i]["StockID"])
        table["股名"].append(data[i]["StockName"])
        table["漲跌幅"].append(data[i]["UpRate"])
        table["成交量"].append(data[i]["DealVol"])
```

# 定義視覺化函式graph()

```
def graph(x, y, z, title):
    fig,ax = plt.subplots()
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
    plt.title(title)
    ax.plot(x, z, color="red", marker="o")
    # set x-axis label
    ax.set_xlabel("公司名",fontsize=14)
    # set y-axis label
    ax.set_ylabel("UpRate(%)",color="red",fontsize=14, loc='center')

    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(x, y, color="blue",marker="o")
    ax2.set_ylabel("DealVol(張)",color="blue",fontsize=14, loc='center')
    plt.show()

# A
xa = df["股名"][0:15]
ya = df["成交量"][0:15]
za = df["漲跌幅"][0:15]
graph(xa,ya,za,titleA)

# B
xb = df["股名"][15:30]
yb = df["成交量"][15:30]
zb = df["漲跌幅"][15:30]
graph(xb, yb, zb, titleB)

# C
xc = df["股名"][30:45]
yc = df["成交量"][30:45]
zc = df["漲跌幅"][30:45]
graph(xc, yc, zc, titleC)

# D
xd = df["股名"][45:60]
yd = df["成交量"][45:60]
zd = df["漲跌幅"][45:60]
graph(xd, yd, zd, titleD)

# E
xe = df["股名"][60:75]
ye = df["成交量"][60:75]
ze = df["漲跌幅"][60:75]
graph(xe, ye, ze, titleE)

# F
xf = df["股名"][75:90]
yf = df["成交量"][75:90]
zf = df["漲跌幅"][75:90]
graph(xf, yf, zf, titleF)
```
